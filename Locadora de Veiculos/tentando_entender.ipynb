```python
def pagina_principal ():
  mensagem_abertura = "Bem-vindo a Locadora de Carros LTDA"
  print("-"*30)
  print(mensagem_abertura)
  print("-"*30,"\n")
  print("O que você precisa para hoje?\n")
  print("1 - Mostrar carros disponíveis\n2 - Alugar um carro\n3 - Devolver um carro")
  print("\n\n\n\n\n\n\n\n")
  print("="*30)
  print('0- Sair')
  main(True)

def exibir_carros():
  print("Os carros disponíveis são:\n")
  for index, carro in enumerate(carros_disponiveis, start=1):
    print(f"{index - 1}- {carro[0]} - R$ {carro[1]} / dia")

def exibir_carros_alugados():
  if len(carros_alugados) == 0:
    print("Não há carros alugados")
    return
  else:
    print("Qual carro deseja devolver :\n")
    for index, carro in enumerate(carros_alugados, start=1):
      print(f"{index - 1}- {carro[0]} - R$ {carro[1]} / dia")

def mensagem():
  print("="*30)
  print('0- Sair')
  print("1- Retornar ao menu principal")

def mostrar_carros(selecao):
  exibir_carros()
  print("O que deseja fazer?\n")
  mensagem()
  while True:
    retornar = int(input())
    if retornar == 1:
      pagina_principal()
    elif retornar == 0:
      break

def alugar_carro(selecao):
  exibir_carros()
  print("Qual carro você deseja alugar?\n")
  carro = int(input())
  print(f"Você escolheu o {carros_disponiveis[carro][0]}, ótima escolha!")
  print(f"O preço dele é R$ {carros_disponiveis[carro][1]} por dia")
  print("Quantos dias você deseja alugar?\n")
  dias = int(input())
  print(f"O total a pagar é de R$ {carros_disponiveis[carro][1] * dias}")
  print("Poderia confirmar o aluguel?\n")
  print("1- Sim\n2- Não")
  while True:
    confirmar = int(input())
    if confirmar == 1:
      carros_alugados.append(carros_disponiveis[carro])
      carros_disponiveis.pop(carro)
      print("Parabéns, você alugou um carro!")
      print("O que deseja fazer?\n")
      mensagem()
      while True:
        retornar = int(input())
        if retornar == 1:
          pagina_principal()
        elif retornar == 0:
          break
    elif confirmar ==2:
      print("Digite 1 para ir para o menu principal ou 2 para sair")
      retornar = int(input())
      if retornar == 1:
        pagina_principal()
      elif retornar == 2:
        break
    else:
      print("Opção inválida, por favor, escolha uma das 2 opções")
  mensagem()

def devolver_carro(selecao):
  exibir_carros_alugados()
  carro = int(input())
  print(f"Você escolheu o {carros_alugados[carro][0]}, obrigado por devolver!")
  print("O que deseja fazer?\n")
  mensagem()
  while True:
    retornar = int(input())
    if retornar == 1:
      pagina_principal()
    elif retornar == 0:
      break

carros_disponiveis =[["Chevrolet Tracker", 120],["Chevrolet Onix", 90],["Hyundai HB20", 85],["Hyundai Tucson", 120],["Fiat Uno", 60],["Fiat Mobi", 70],["Fiat Pulse", 130]]

carros_alugados = []

def main (x):
    while x:
      selecao = int(input())
      if selecao < 0 or selecao > 3:
        print("Opção inválida, por favor, escolha uma das 3 opções")
      else:
        if selecao == 1:
          mostrar_carros(selecao)
        elif selecao == 2:
          alugar_carro(selecao)
        elif selecao == 3:
          devolver_carro(selecao)
        elif selecao == 0:
          break

pagina_principal()
```
